# 4. Find the Most Common Accident Location
 Task: Identify the most frequent accident location.

```python
import pandas as pd                                              # Import pandas library for data manipulation
df = pd.read_csv("road.csv")                                     # Read the CSV file 'road.csv' into a DataFrame


most_common_location = df['Total number of Fatal Accidents - 2011'].value_counts().idxmax()  # Find the value that occurs most often in the 'Total number of Fatal Accidents - 2011' column

print("Most common accident location:", most_common_location)    # Print the most common value found

```

# 5 Find the Average Number of Casualties
 Task: Compute the average number of casualties per accident.

```python
import pandas as pd                                                  # Import pandas library to work with data
df = pd.read_csv("road.csv")                                         # Read the CSV file 'road.csv' into a DataFrame


average_casualties = df['Severity - 2015'].mean()                    # Calculate the mean (average) value of the 'Severity - 2015' column

print("Average number of Severity - 2015 per accident:", round(average_casualties, 2))  # Print the average rounded to 2 decimal places
```

# 6 Filter Data for High-Speed Accidents
 Task: Extract accidents where the vehicle speed was more than 80 km/hr.

```python
import pandas as pd                                                   # Import the pandas library to handle data
df = pd.read_csv("road.csv")                                          # Load the CSV file named 'road.csv' into a DataFrame


high_speed_accidents = df[df['Minor Accidents - 2014'] > 80]          # Select rows where 'Minor Accidents - 2014' value is greater than 80


print(high_speed_accidents.head())                                    # Show the top 5 rows from the filtered DataFrame

```

# 7. Identify the Deadliest Road
 Task: Find the road with the highest average casualties.

```python
import pandas as pd                                                                 # Import pandas for data manipulation
df = pd.read_csv("road.csv")                                                        # Load the CSV file into a DataFrame


deadliest_road = df.groupby('Severity - 2013')['Persons Killed - 2011'].mean().idxmax()  # Group by 'Severity - 2013', calculate mean of 'Persons Killed - 2011', and get the group with the highest average

display("The deadliest road (highest average Persons Killed - 2011):", deadliest_road)   # Display the result

```

# 8. Filter Records with More Than 5 Fatalities

# 9. Find Maximum and Minimum Number of People Injured

```python
import pandas as pd                                                            # Import pandas library to work with data
df = pd.read_csv("road.csv")                                                  # Load the CSV file 'road.csv' into a DataFrame

max_injured = df['Total Accidents - 2014'].max()                              # Find the maximum value in 'Total Accidents - 2014' column
min_injured = df['Total Accidents - 2014'].min()                              # Find the minimum value in 'Total Accidents - 2014' column

display("Maximum Total Accidents - 2014 injured:", max_injured)               # Display the maximum injury count
display("Minimum Total Accidents - 2014 injured:", min_injured)               # Display the minimum injury count
```

# 11. Bar Chart – Accidents by City

```python
import pandas as pd                                              # Import pandas for data handling
import matplotlib.pyplot as plt                                  # Import matplotlib for plotting

df = pd.read_csv("road.csv")                                     # Load data from 'road.csv'

city_counts = df["Name of City"].value_counts().head(10)         # Get top 10 cities with most accidents

city_counts.plot(kind='bar', color='skyblue')                    # Create a bar chart with sky blue color
plt.title("Top 5 Cities with Most Accidents")                    # Set the chart title
plt.xlabel("City")                                               # Label the X-axis
plt.ylabel("Number of Accidents")                                # Label the Y-axis
plt.xticks(rotation=45)                                          # Rotate city names for better visibility
plt.tight_layout()                                               # Adjust layout to fit everything nicely

plt.show()                                                       # Show the plot
```

# 12. Pie Chart – Accident Distribution by Road Type

```python
import pandas as pd                                        # Import pandas for data handling
import matplotlib.pyplot as plt                            # Import matplotlib for plotting

df = pd.read_csv("road.csv")                               # Read the CSV file into a DataFrame

road_type_counts = df["Severity - 2014"].value_counts()    # Count how many times each severity appears in 2014

# Create a pie chart to show distribution
plt.pie(road_type_counts, labels=road_type_counts.index, autopct='%1.1f%%', startangle=90)

plt.title("Accident Distribution by Road Type")            # Set title for the chart
plt.axis('equal')                                          # Make the pie chart a perfect circle
plt.show()                                                 # Display the pie chart
```

# 13. Line Chart – Accidents per year

```python
import pandas as pd                            # Import pandas for data handling
import matplotlib.pyplot as plt                # Import matplotlib for plotting

df = pd.read_csv("road.csv")                   # Read the CSV file into a DataFrame

# Count the number of accidents for each unique value in 'Total Accidents - 2015' and sort by the value
yearly_accidents = df['Total Accidents - 2015'].value_counts().sort_index()

# Create a line plot for accidents in 2015
plt.plot(yearly_accidents.index, yearly_accidents.values, marker='o', color='blue')

plt.title("Accidents Year - 2015")             # Set the title of the chart
plt.xlabel("Year")                             # Set the x-axis label
plt.ylabel("Number of Accidents")              # Set the y-axis label
plt.grid(True)                                 # Add gridlines to the chart
```

# 15.Scatter Plot – Name of City vs. Killed - 2015

# 16. TWO GRAPH IN ONE PROGRAM


```python
import pandas as pd                                # Import pandas for data handling
import matplotlib.pyplot as plt                    # Import matplotlib for plotting

df = pd.read_csv("road.csv")                       # Load the CSV file

accident_by_city = df['Name of City'].value_counts().head(5)   # Get top 5 cities with most accidents

weather_conditions = df['Severity - 2015'].value_counts()      # Count severity cases in 2015


plt.figure(figsize=(12, 5))                        # Set figure size for side-by-side charts

plt.subplot(1, 2, 1)                               # First plot (Bar Chart) on left side
accident_by_city.plot(kind='bar', color='blue') # Bar chart for top cities
plt.title("Top 5 Cities with Most Accidents")      # Title for bar chart
plt.xlabel("City")                                 # X-axis label
plt.ylabel("Number of Accidents")                  # Y-axis label

plt.subplot(1, 2, 2)                               # Second plot (Pie Chart) on right side
plt.pie(weather_conditions, labels=weather_conditions.index,
        autopct='%1.1f%%', startangle=90)          # Create pie chart with percentage
plt.title("Accidents Severity - 2015")             # Title for pie chart
plt.gca().set_aspect('equal')                      # Make pie chart round

plt.tight_layout()                                 # Adjust spacing between plots
plt.show()                                         # Show both charts
```
